```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from sklearn.metrics import classification_report, confusion_matrix 
```

# Carregando os dados de entrada a partir do arquivo .csv

```python
inputs = pd.read_csv("/content/sample_data/X_In.csv")
inputs.head()



```

# Carregando os dados de saídas esperadas a partir do arquivo .csv


```python
outputs = pd.read_csv("/content/sample_data/Y_Out.csv")
outputs.head()
```

#Aqui é feita a classificação de 20% da entrada como arquivo de teste e separando os arquivos.

```python
input_train, input_test, output_train, output_test = train_test_split(inputs,outputs,test_size=0.2, random_state=36)

input_train.head()
```

#Definindo a estrutura da rede neural. Isso inclui o número de camadas e o número de neurônios em cada camada. Como também a escolha das funções de ativação.



```python
model = tf.keras.Sequential()

# Adicione uma camada de entrada com 7 neurônios, uma camada escondida com 14 neurônios
# e uma camada de saída com 4 neurônios
model.add(tf.keras.layers.InputLayer(input_shape=(7,)))
model.add(tf.keras.layers.Dense(7, activation="tanh"))
model.add(tf.keras.layers.Dense(15, activation="tanh"))
model.add(tf.keras.layers.Dense(4, activation="sigmoid"))

model.summary()
```

#Compilando o modelo especificando o otimizador e a função de perda a serem usados.

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['binary_accuracy'])
```

#Treinando o modelo usando o método fit e fornecendo o conjunto de dados de treinamento e as configurações de treinamento.


```python
model.fit(input_train, output_train, epochs=200, batch_size=15)
```

#Fazendo uma comparação entre os valores previstos e valores esperados.

```python
output_pred = model.predict(input_test)

passed = 0
failed = 0

output_test = output_test.to_numpy()

print(type(output_pred))

print(type(output_test))

for i in range(len(input_test)):
  if(np.array_equal(np.round(output_pred[i]).astype(int),output_test[i])):
    print(np.round(output_pred[i]).astype(int),"==========", output_test[i],'PASSED')
    passed+=1
  else:
    print(np.round(output_pred[i]).astype(int),"==========", output_test[i],'FAILED')
    failed+=1

print("Passed:", passed)
print("Failed:", failed)



```

Obtenha as entradas que você deseja testar do DataFrame. 


```python
loss, accuracy = model.evaluate(inputs, outputs)

print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```
